```python
import pandas as pd

# CSV 파일 불러오기
df = pd.read_csv('../data/BD_REGIST_no_remove3000.csv', encoding='cp949')

# 중복 행 개수 확인
num_duplicates = df.duplicated().sum()
print(f"중복 행 개수: {num_duplicates}개")

# 중복 제거
df_no_duplicates = df.drop_duplicates()

# 중복 제거 후 데이터프레임 크기 확인
print(f"중복 제거 후 데이터 크기: {df_no_duplicates.shape}")

# 필요 시 새로운 파일로 저장
# df_no_duplicates.to_csv('BD_REGIST_no_duplicates.csv', index=False)
```

```python
# BD 데이터프레임에서 MULTI_YN == 'N' 필터링
bd_df_n = df[df['MULTI_YN'] == 'N'].copy()

# 필터링된 데이터 확인 (선택적)
print("MULTI_YN == 'N' 데이터프레임 개수:", len(bd_df_n))

# RECAP_PK 중복 여부 확인
recap_duplicates = bd_df_n['RECAP_PK'].duplicated(keep=False)  # 중복된 모든 행 True
duplicated_recap_rows = bd_df_n[recap_duplicates]

# 중복이 있는 경우 출력
if not duplicated_recap_rows.empty:
    print("중복된 RECAP_PK 목록:")
    print(duplicated_recap_rows[['RECAP_PK']].drop_duplicates())
    print("\n중복된 행:")
    print(duplicated_recap_rows)
else:
    print("중복된 RECAP_PK가 없습니다.")
```
