```python
import pandas as pd

# 엑셀 파일 읽기
df_1 = pd.read_excel("../data/text_matching_result_ver3.xlsx")

# 필터링할 MATCH_STAGE 값 리스트
filter_values = [
    '1차', 
    '2차', 
    '3차', 
    '4차(DONG일치)', 
    '수동매칭(1차중복)', 
    '수동매칭(2차중복)', 
    '수동매칭(DONG일치)'
]

# MATCH_STAGE 컬럼의 값이 filter_values 리스트에 포함된 행만 필터링
filtered_df = df_1[df_1['MATCH_STAGE'].isin(filter_values)]

# 결과 확인
filtered_df

```

```python
import pandas as pd

# 1. 에너지 데이터에서 필요한 컬럼 리스트 (건축물ID 포함)
energy_vars = [
    '단위면적당 1차 E사용량_전기 (KWH)',
    '단위면적당 1차 E사용량_가스 (KWH)',
    '단위면적당 1차 E사용량_지역냉난방 (KWH)',
    '단위면적당 1차 E사용량_유류 (KWH)',
    '단위면적당 1차 E사용량_기타 (KWH)',
    'EUI(1차E/냉난방 면적)',
    'EUI(1차E/연면적)',
    '준공년도',
    'E사용량 (kWh)_전기 (KWH)',
    'E사용량 (kWh)_가스 (KWH)',
    'E사용량 (kWh)_지역냉난방 (KWH)',
    'E사용량 (kWh)_유류 (KWH)',
    'E사용량 (kWh)_기타 (KWH)',
    'E사용량 (toe)_전기 (toe)',
    'E사용량 (toe)_가스 (toe)',
    'E사용량 (toe)_지역냉난방 (toe)',
    'E사용량 (toe)_유류 (toe)',
    'E사용량 (toe)_기타 (toe)',
    '온실가스 배출량(tCO2)_전기',
    '온실가스 배출량(tCO2)_가스',
    '온실가스 배출량(tCO2)_지역난방',
    '온실가스 배출량(tCO2)_유류',
    '온실가스 배출량(tCO2)_기타',
    '온실가스 배출량(tCO2)_총계'
]

# 에너지 데이터에서 '건축물ID'와 위 변수들만 선택
energy_subset = df[['건축물ID'] + energy_vars]

# 2. filtered_df (텍스트 매칭 결과)와 energy_subset을 건축물ID 기준으로 왼쪽 병합(merge)
merged_df = pd.merge(filtered_df, energy_subset, left_on='SEQ_NO', right_on='건축물ID', how='left')

# 3. 병합 결과에서 에너지 데이터가 붙은 건수를 파악합니다.
# 에너지 데이터가 붙지 않은 경우, energy_vars에 해당하는 컬럼들은 모두 NaN으로 남게 됩니다.
# 여기서는 첫번째 에너지 변수('단위면적당 1차 E사용량_전기 (KWH)')가 NaN이 아닌 경우를 매칭된 것으로 판단합니다.
matched_count = merged_df['단위면적당 1차 E사용량_전기 (KWH)'].notna().sum()
total_count = merged_df.shape[0]
unmatched_count = total_count - matched_count

print(f"총 {total_count}건 중, {matched_count}건이 건축물ID 기준으로 에너지 데이터와 매칭되었습니다.")
print(f"{unmatched_count}건은 건축물ID가 없어 에너지 데이터와 매칭되지 않았습니다.")
```

```python
# 엑셀 파일로 저장
merged_df.to_excel('../result/merged_energy_data.xlsx', index=False)
print("엑셀 파일이 성공적으로 저장되었습니다.")
```
