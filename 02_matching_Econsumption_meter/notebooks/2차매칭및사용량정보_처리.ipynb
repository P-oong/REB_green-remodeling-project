```python
import pandas as pd
import os

# 결과 디렉토리 경로
result_dir = r"..\result"

# 사용량정보 CSV 파일 경로
usage_file = os.path.join(result_dir, "사용량정보.csv")

# CSV 파일 읽기 (헤더 없이)
df = pd.read_csv(usage_file, header=None)

# 이미지에 나와있는 순서대로 컬럼명 부여
column_names = [
    '공공건축물_PK',
    '에너지공급기관코드',
    '에너지고유번호',
    '사용년월',
    '사용량일련번호',
    '사용시작일자',
    '사용종료일자',
    '사용용도코드',
    '사용량',
    '단위코드'
]

# 컬럼명 지정
df.columns = column_names

# 데이터 확인
print(f"데이터 크기: {df.shape}")
print("\n처음 5개 행:")
print(df.head())

# 수정된 파일 저장
df.to_csv(usage_file, index=False, encoding='utf-8-sig')
print(f"\n컬럼명이 업데이트된 파일이 저장되었습니다: {usage_file}")
```

```python
df_use = pd.read_csv(r"..\result\사용량정보.csv")
df_match = pd.read_csv(r"..\result\매칭정보.csv")
```

```python

df_use = df_use.sort_values(by='공공건축물_PK')
df_match = df_match.sort_values(by='공공건축물_PK')
```

```python
col_name = '에너지고유번호'

# 각 값의 빈도를 계산하여 중복(빈도수 > 1)인 값들만 추출합니다.
duplicates = df_use[col_name].value_counts()[df_use[col_name].value_counts() > 1]

print("중복된 값 및 각각의 개수:")
print(duplicates)

# 전체 중복 행의 수: 각 행이 중복인지 여부를 체크 (첫 등장 제외)
total_duplicates = df_use[col_name].duplicated().sum()
print(f"\n총 중복 행의 수: {total_duplicates}건")
```
